- NTFS - both windows and linux
- EXT4 - linux only
- FAT32 - linux, windows, mac - not support files larger that 4GB, size of file system can't be larger than 32GB

## Partition
- sub disk iniside physical disk
## Volume
- file system inside the partition
## Partition table
- table that hold the partition details 
- MBR (master boot record), GPT (GUID partition table)
### MBR
- volume size of 2 TB or less
- only can have 4 primary partition
- if you want more partition, take the primary and make it into extended partition. inside the extended partition, make it into logical partition
### GPT
- volume size > 2TB
- 1 type of partition

- `DiskPart` windows
- `parted -l` ubuntu







## windows files
- MFT (master files table) - record the metadata of each file
- shortcut - shortcut to a file, when open the file, it will not open the actual file.
- symllink - like shortcut but when try to open it, it will open the actual file.
- `mklink <file_sym> <original_file>`
- `mklink /H <file_sym> <originall_file>` hard link, point to MFT of that file, if the original file is rename, the syslink will still can point to the original file

## linux files
- metadata of file stored in inodes
- soflink - symbolic link
    - `ln -s <file>`
- hard link 


- `du -h <folder>`
- `df -h <folder>`
