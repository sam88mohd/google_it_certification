## Windows Powershell
`Get-Help`  
`Get-Alias`  
`Get-ChildItem C:\`  
`sls <string> <fiilename>` or `Select-String <string> <filename>`  
`ls C:\ -Recurse -Filter <file>`

`echo woof > dog.txt`  
`rm service 2> $null` Filter out error messsage  
`Get-Help about_redirection`  
`Get-LocalUser`  
`Get-LocalGroup`  
`Get-LocalGroupMember <groupname>`  
`net user <user> /logonpasswordchg:yes` // change user password  
`net user <user> /add /logonpasswordchg:yes`  
`net user <user> /delete`

## Linux
`rm service > /dev/null` Filter out error message  
`mkdir -p -m -v`  -p allow to create parent, -m set permission  
`passwd -e <user>` force user to change password next time he login  
`sudo useradd <user>`  
`sudo userdelete <user>`
