## WIndows File Permissions
- ACL - Access COntrol List  
- DACL - Discretionary Access COntrol List - as a note about who can use the file and what they allowed to do with it  
- SACL - System Access Control List - tell windows to use event log everytime someone access a file or folder 










- to view and change ACL:
    - `icacls <file>`
    - `icacls /?` show helps
    - `icacls '<file path>' /grant 'Everyone:(OI)(CI)(R)'`
    - `icacls '<file path>' /remove Everyone`
- Special permission


## Linux FIle Permission

- first - is file, dr is directory
- -rwx - file permission of the owner
- rw- - file permission of group
- r-- file permisson of all other users

`chmod u+x <file>`  
 - u for owner
 - g for group
 - o other user
 
 `chmod 777 <file>`
 - r,w,x 7
 - r 4
 - w 2
 - x 1
 - \- 0
 
 - Special Permission
     - setuid // run root priveledge file without sudo
     
     - setgid // enable the file to run as group member
     
     - sticky bit // make anyone can write to a file or folder but cant delete anything. only owner can delete the content
     

`chown <user>` change owner
