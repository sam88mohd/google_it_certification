## Managing service in linux
- `sudo service --status-all` list all service

- `sudo service <servicename> status`
- `logger This is a test log entry` this will write in syslog
- `sudo tail -1 /var/log/syslog`
- `sudo service <servicename> stop`
- `sudo service <servicename> start`
- `sudo ls -l /etc/<service folder>` check content of the service
- `sudo ls -l /var/log/cups` log created by cups
- `sudo service cups restart`
- `sudo service cups reload`

## Managing service in windows
- `Get-Service`
- `Get-Service <servicename>`
- `Get-Service wisvc | Format-List *` get more info
- `Start-Service wisvc`
- `Stop-Service wisvc`
- `Set-Service ScardSvr -StartupType Manual`
- `Install-WindowsFeature Web-WebServer,Web-Mgmt-Tools -IncludeAllSubFeature` install feature not installed by default
