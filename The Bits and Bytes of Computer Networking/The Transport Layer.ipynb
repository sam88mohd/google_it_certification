## TCP

- tcp will split the data into multiple segment- sequence number determine the order of the segment
- Data offset field - a 4 bit number that communicates how long the TCP header for the segment is
- TCP window - specifies the range of sequence number that might be sent before an acknoledgment is required.

- The way TCP established a connections is through the use of different TCP control flags used in a very specific order

## Flags

### URG - Urgent
- urgent data pointer field has more data about this.
### ACK - Acknowledged
- value of one in this field means that the acknowledgement number field should be examine. 
### PSH - Push
- the transmitting device want the receiving device to push currently-buffered data to the application on the receiving end as soon as possible
### RST - reset
- one of the side in a TCP connection hasn't been able to properly recover from a series of missing or malformed segments
- bassiclally to start over again if the data received is corrupted
### SYN - Synchronize
- it is used when first establishibing a TCP connection and make sure the receiving end knows to examine the sequence number flags
### FN - finish
- when this flag set to 1, it meand the transmitting computer doesn't have any data to send and the connection can be closed.

## TCP Socket states
- A socket is the instantiation of an endpoint in a potential TCP connection
### LISTEN
- socket is ready and listening for incoming connections.
- server side only
### SYN_SENT
- a sychronization request has been sent, but the connection hasn't been established yet
- client side only
### SYN-RECEIVED
- this mean a socket previously in a LISTEN state has received a synchronization request and sent a SYN/ACK back but ut hasn't received the final ACK from the client yet.
- server side only
### ESTABLISHED
- this mean TCP connection is in wirking order
- both server and client side.
### FIN_WAIT
- this mean a FIN has been sent but the corresponding ACK from the other end hasn't been recieved yet.
### CLOSE_WAIT
- this means that connection has been closed at the TCP layer, but the application that opened the socket hasn't release on the socket yet.
### CLOSED
-  this means that the connection has been fully terminated and that no further communication is possible
