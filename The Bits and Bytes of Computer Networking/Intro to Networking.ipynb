# Intro to Networking

## TCP/IP Five Layer 

*TCP/IP Five Layer Network Model*  
1. physical  
- include specification for the networking cables and the conectectors that join devices together along with specifications scribing how signal are sent over these connection.  
- delivery truck and roads.  
2. data link  
- responsible for defining a common way of interpreting these signals so network device can communicate.  
- how the delivery truck get from one intersection to the next over and over.  
3. network  
- this layer alow different networks to communicate with each other through device knowns as routers. responsible for getting data delivers accross a collection of networks.  
- identifies which roads need to be takej to get from address A to address B.  
4. transport  
- sort out which client and server programs are supposed to get the data.
- ensures that delivery driver knows how to knoc on your door to tell you your package has arrived.  
5. application
- application specific 
- content of the package itself.  

## Cables

crosstalk - when electical pulse on 1 wire is accidently detected on another wire

- cat5 - older cable, been replaced by cat5e and cat6  
- cat5e - more reliable, better to prevent crosstalk and prevent data need to retransmit
- cat6 - more speed and reliable. cons is have shorter max distance when used at higher speeds.
- fiber - use optic (light)


## Hub and switches
### Hub

collision domain - a network segment where only one device can communicate at a time

### Switch

a layer 2 device - data link layer - can inspect content of the data transmit to determine the destination

### Router

a layer 3 device - inspect IP data tp determine where to send things. router store internal tables containing information about how to route traffic between lots of diferent networks all over the world.  

router share data with eachouter via BGP - border gateway protocol - get optimal path to foward traffic

## IP Address
- ip address (32 bit) 4 octet 
packet in network layer call ip datagram  





ip address can be split into 2 section:
- network ID - identify network id
- host ID - identify individual host

example:  



### Class A
- first octet is network ID, other 3 is Host ID

### Class B
- first, second octet is network ID, 2 is Host ID

### Class C
- first, second, third octet is network ID, last is Host ID

## ARP - Address resolution protocol

protocol used to discover the hardware address of a node with a certain IP address - how is IP is relate to each other  

### ARP table
- a list of IP addresses and the MAC addresses associated with them




## Subnetting

- process of taking a large network and splitting it up into many individual and smaller subnetworks or subnets



- part with 0 in subnet mask is the one to keep when determine the host id  - by using the IP class, we know that the IP range is Class A. so 1 octet is not count. the remaining octet with 1 in subnet mask is the Subnet ID.
- the octet with 0 in subnet mask is the host ID.

- subnet can only have 2 less than 255 to be assigned.
- 0 not use, 255 is reserved as a broadcast address for the subnet.


- 9.100.100.100/27 refer to 1 in subnetmask

## CIDR - classles inter-domain routing
- in CIDR, network ID and subnet ID is combine to 1 to become network ID

*9.100.100.100/24 - example of CIDR*
