## Names Resolution
### Domain Name System - DNS
- DNS turn IP addresses into readable format
- DNS help administrate network
- DNS is an application layer that use UDP
#### Name resolution
- process of DNS resolve domain name into an IP address
#### 5 primary types of DNS server:
1. Caching name servers
2. Recursive name servers
3. Root name servers - total 13
4. TLD name servers - top level domain (.com)
5. Authoritives name servers - actual server 



## Record types
### A record 
- used to point a certain domain name at a certain IPV4 
- 1 A record only for 1 domain name
- 1 domain name can have multiple A record
### Quad A record
- same as A record but return IPV6
### CNAME record
- route a domain to another domain - example WWW.microsoft.com, microsoft.com will point in same IP address
### MX exchange (MX)
- used in order to deliver email to the correct server
### Service record (SRV)
- used to define the location of various specific services
